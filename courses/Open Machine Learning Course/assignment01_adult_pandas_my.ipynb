```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('../../data/adult.data.csv')
data.head()
```

**1. Сколько мужчин и женщин (признак *sex*) представлено в этом наборе данных?**

```python
data[data['sex']=='Female'].count()
```

**2. Каков средний возраст (признак *age*) женщин?**

```python
data[data['sex']=='Female']['age'].mean()
```

**3. Какова доля граждан Германии (признак *native-country*)?**

```python
data[data['native-country']=='Germany']['native-country'].count()
data[data['native-country']=='Germany']['native-country'].count()/data['native-country'].count()*100
```

**4-5. Каковы средние значения и среднеквадратичные отклонения возраста тех, кто получает более 50K в год (признак *salary*) и тех, кто получает менее 50K в год? **

```python
data[data['salary']=='<=50K']['age'].describe()
```

```python
columns_to_show = ['age']

data.groupby(['salary'])[columns_to_show].agg([np.mean, np.std])
```

**6. Правда ли, что люди, которые получают больше 50k, имеют как минимум высшее образование? (признак *education – Bachelors, Prof-school, Assoc-acdm, Assoc-voc, Masters* или *Doctorate*)Неверно!!!!**

```python
pd.crosstab(data['salary'], data['education'])
```

**7. Выведите статистику возраста для каждой расы (признак *race*) и каждого пола. Используйте *groupby* и *describe*. Найдите таким образом максимальный возраст мужчин расы *Amer-Indian-Eskimo*.**

```python
columns_to_show = ['age']

data.groupby(['race'])[columns_to_show].agg([np.mean, np.std, np.max])
```

**8. Среди кого больше доля зарабатывающих много (>50K): среди женатых или холостых мужчин (признак *marital-status*)? Женатыми считаем тех, у кого *marital-status* начинается с *Married* (Married-civ-spouse, Married-spouse-absent или Married-AF-spouse), остальных считаем холостыми.(Неверно!!!!)**

```python
print(data[((data['marital-status'] == 'Married-civ-spouse') | (data['marital-status'] == 'Married-spouse-absent') | (data['marital-status'] == 'Married-AF-spouse') )][data['salary']=='>50K']['salary'].count())
print(data[((data['marital-status'] != 'Married-civ-spouse') | (data['marital-status'] != 'Married-spouse-absent') | (data['marital-status'] != 'Married-AF-spouse') )][data['salary']=='>50K']['salary'].count())
```

**9. Какое максимальное число часов человек работает в неделю (признак *hours-per-week*)? Сколько людей работают такое количество часов и каков среди них процент зарабатывающих много?**

```python
print(data['hours-per-week'].max())
print(data[data['hours-per-week']==99]['age'].count())
(data[data['hours-per-week']==99][data['salary']=='>50K']['age'].count())/(data[data['hours-per-week']==99]['age'].count())
```

**10. Посчитайте среднее время работы (*hours-per-week*) зарабатывающих мало и много (*salary*) для каждой страны (*native-country*).**

```python
columns_to_show = ['hours-per-week']

data.groupby(data[data['salary']=='>50K']['native-country'])[columns_to_show].agg([np.mean])
```
